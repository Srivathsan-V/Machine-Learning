## Logistic Regression

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
categorical_columns = list(datacopy.dtypes[datacopy.dtypes == 'object'].index)
numerical_columns = list(datacopy.dtypes[datacopy.dtypes != 'object'].index)

numerical_columns.remove('median_house_value')
```

```python
dv = DictVectorizer(sparse=False)

train_dict = df_train[categorical_columns + numerical_columns].to_dict(orient='records')
X_train = dv.fit_transform(train_dict)

val_dict = df_val[categorical_columns + numerical_columns].to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver="liblinear", C=1.0, max_iter=1000, random_state=42)
model.fit(X_train,y_train)
```

```python
y_pred = model.predict_proba(X_val)[:,1]
```

```python
y_pred
```

```python
median_house_value = (y_pred >= 0.5)
median_house_value
```

```python
df_pred = pd.DataFrame()
df_pred['probability'] = y_pred
df_pred['prediction'] = median_house_value.astype(int)
df_pred['actual'] = y_val
```

```python
df_pred['correct'] = df_pred.prediction == df_pred.actual
```

```python
accuracy = df_pred.correct.mean().round(2)
accuracy
```
